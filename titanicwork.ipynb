```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.impute import SimpleImputer

class AgeImputer(BaseEstimator, TransformerMixin):
    
    def __init__(self):
        # The constructor is empty because we don't have any hyperparameters to set
        pass
        
    def fit(self, X, y=None):
        # The fit method is required for compatibility with scikit-learn transformers.
        # However, since we don't need to learn anything from the data for imputation,
        # we simply return self (the instance of the class).
        return self
        
    def transform(self, X):
        # The transform method performs the actual imputation on the 'Age' column of X.

        # Step 1: Create a SimpleImputer instance with the strategy "mean".
        imputer = SimpleImputer(strategy="mean")

        # Step 2: Use the SimpleImputer to compute the mean of the 'Age' column (if missing) 
        # and replace the missing values in 'Age' with the computed mean.
        X['Age'] = imputer.fit_transform(X[['Age']])

        # Step 3: Return the updated DataFrame with the imputed 'Age' column.
        return X
from sklearn.preprocessing import OneHotEncoder

class FeatureEncoder(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        return self 
    
    def transform(self, X):
        encoder = OneHotEncoder()
        matrix = encoder.fit_transform(X[['Embarked']]).toarray()
        
        column_names = ["C", "S", "Q", "N"]
        
        for i in range(len(matrix.T)):
            X[column_names[i]] = matrix.T[i]
        
        matrix = encoder.fit_transform(X[['Sex']]).toarray()
        column_names = ["Female", "Male"]
        for i in range(len(matrix.T)):
            X[column_names[i]] = matrix.T[i]
        
        return X

class FeatureDropper(BaseEstimator,TransformerMixin):
    
    def fit(self,X,y=None):
        return self
    def transform(self,X):
        return X.drop(["Embarked","Name","Ticket","Cabin","Sex","N"],axis=1,errors="ignore")

    
from sklearn.pipeline import Pipeline
pipeline=Pipeline([("ageimputer",AgeImputer()),
                   ("featureencoder",FeatureEncoder()),
                   ("featuredropper",FeatureDropper())])
strat_train_set=pipeline.fit_transform(strat_train_set)
strat_train_set

            
```
